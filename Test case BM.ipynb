```python
import pandas as pd
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String, Date, Float, BigInteger
import os
```

```python
engine = create_engine('sqlite:///subscriptions.db')
metadata = MetaData()
```

```python
subscriptions = Table('subscriptions', metadata,
                      Column('EventDate', Date),
                      Column('AppName', String),
                      Column('AppAppleID', BigInteger),
                      Column('SubscriptionName', String),
                      Column('SubscriptionAppleID', BigInteger),
                      Column('SubscriptionGroupID', BigInteger),
                      Column('SubscriptionDuration', String),
                      Column('IntroductoryPriceType', String),
                      Column('IntroductoryPriceDuration', String),
                      Column('MarketingOptInDuration', String),
                      Column('CustomerPrice', Float),
                      Column('CustomerCurrency', String),
                      Column('DeveloperProceeds', Float),
                      Column('ProceedsCurrency', String),
                      Column('PreservedPricing', String),
                      Column('ProceedsReason', String),
                      Column('Client', String),
                      Column('Device', String),
                      Column('Country', String),
                      Column('SubscriberID', BigInteger),
                      Column('SubscriberIDReset', String),
                      Column('Refund', String),
                      Column('PurchaseDate', Date),
                      Column('Units', Float))

metadata.create_all(engine)

print("Таблиця 'subscriptions' була успішно створена.")
```

```python
folder_path = 'itunes_dataset'  

all_data = pd.DataFrame()

for filename in os.listdir(folder_path):
    if filename.endswith('.txt'):  
        file_path = os.path.join(folder_path, filename)
        data = pd.read_csv(file_path, sep='\t')
        all_data = pd.concat([all_data, data], ignore_index=True)
```

```python
all_data.rename(columns={
    'Event Date': 'EventDate',
    'App Name': 'AppName',
    'App Apple ID': 'AppAppleID',
    'Subscription Name': 'SubscriptionName',
    'Subscription Apple ID': 'SubscriptionAppleID',
    'Subscription Group ID': 'SubscriptionGroupID',
    'Subscription Duration': 'SubscriptionDuration',
    'Introductory Price Type': 'IntroductoryPriceType',
    'Introductory Price Duration': 'IntroductoryPriceDuration',
    'Marketing Opt-In Duration': 'MarketingOptInDuration',
    'Customer Price': 'CustomerPrice',
    'Customer Currency': 'CustomerCurrency',
    'Developer Proceeds': 'DeveloperProceeds',
    'Proceeds Currency': 'ProceedsCurrency',
    'Preserved Pricing': 'PreservedPricing',
    'Proceeds Reason': 'ProceedsReason',
    'Client': 'Client',
    'Device': 'Device',
    'Country': 'Country',
    'Subscriber ID': 'SubscriberID',
    'Subscriber ID Reset': 'SubscriberIDReset',
    'Refund': 'Refund',
    'Purchase Date': 'PurchaseDate',
    'Units': 'Units'
}, inplace=True)
```

```python
all_data
```

```python
all_data.to_sql('subscriptions', engine, if_exists='append', index=False)

print("Дані були успішно завантажені у базу даних.")
```

**Тривалість підписки з заданим id:**

```python
subscription_id = input("Будь ласка, введіть ID підписки: ")

query = f"""
SELECT SubscriptionAppleID, SubscriptionDuration
FROM subscriptions
WHERE SubscriptionAppleID = {subscription_id}
GROUP BY SubscriptionDuration;
"""

result = pd.read_sql_query(query, engine)
result
```

**Список підписок для додатку з заданим id:**

```python
app_apple_id = input("Будь ласка, введіть ID додатку: ")

query = f"""
SELECT AppAppleID, SubscriptionName
FROM subscriptions
WHERE AppAppleID = {app_apple_id}
GROUP BY SubscriptionName;
"""

result = pd.read_sql_query(query, engine)
result
```

**Який додаток приніс більше доходу за заданий період:**

```python
start_date = input("Будь ласка, введіть дату початку у форматі рррр-мм-дд: ")  
end_date = input("Будь ласка, введіть дату закінчення у форматі рррр-мм-дд: ")    

query = f"""
SELECT AppName, SUM(DeveloperProceeds) AS TotalRevenue
FROM subscriptions
WHERE PurchaseDate BETWEEN '{start_date}' AND '{end_date}'
GROUP BY AppName
ORDER BY TotalRevenue DESC
LIMIT 1;
"""
result = pd.read_sql_query(query, engine)
result
```

**Конверсія з оформлення пробного періоду в успішний платіж для користувачів, що зробили підписку id в дату date:**

```python
subscription_id = input("Будь ласка, введіть ID підписки: ")
analysis_date = input("Будь ласка, введіть дату у форматі рррр-мм-дд: ")

query = """
WITH Trials AS (
    SELECT 
        SubscriberID,
        COUNT(*) AS TrialCount
    FROM 
        subscriptions
    WHERE 
        SubscriptionAppleID = '{}'
        AND (EventDate = '{}' OR PurchaseDate = '{}')
        AND IntroductoryPriceType LIKE '%Free Trial%'
    GROUP BY 
        SubscriberID
),
SuccessPayments AS (
    SELECT 
        SubscriberID,
        COUNT(*) AS SuccessCount
    FROM 
        subscriptions
    WHERE 
        SubscriberID IN (SELECT SubscriberID FROM Trials)
        AND CustomerPrice > 0
    GROUP BY 
        SubscriberID
)
SELECT 
    (SUM(SuccessCount) * 100.0) / SUM(TrialCount) AS ConversionRate
FROM 
    Trials
    LEFT JOIN SuccessPayments USING (SubscriberID);
""".format(subscription_id, analysis_date, analysis_date)

result = pd.read_sql_query(query, engine)
conversion_rate_percentage = result["ConversionRate"].values[0]
print(f"Коверсія для підписки {subscription_id} в дату {analysis_date} дорівнює {conversion_rate_percentage:.2f}%")
```
